```python
inmuebles.Valor[0] + inmuebles.Valor[3] + inmuebles.Valor[:5]
```

```python
inmuebles.Valor[0].split()
```

```python
inmuebles.Valor.str.split()
```

```python
valor = inmuebles.Valor.str.split(expand=True)
inmuebles["Moneda"] = valor[0]
inmuebles["Precio"] = valor[1]
inmuebles.sample(3)
```

```python
inmuebles.info()
```

```python
inmuebles["Precio"].sum()
```

```python
inmuebles["Precio"] = inmuebles["Precio"].str.replace(".","",regex=True)
```

```python
inmuebles.info()
```

```python
inmuebles[["Precio","Barrio"]]
```

```python
inmuebles["Precio_Millon"] = inmuebles.Precio.astype("float")/1000000
```

```python
inmuebles.info()
```

```python
inmuebles.Precio_Millon.sum()
```

```python
inmuebles.describe()
```

```python
pd.set_option("display.precision", 2)
pd.set_option("display.float_format", lambda x: "%.2f" % x)
inmuebles.describe()
```

```python
inmuebles["Precio_Millon"].plot.hist(bins=10)
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

plt.figure(figsize=(10,6))
grafica = sns.histplot(data=inmuebles, x="Precio_Millon",kde=True, hue="Tipo")
grafica.set_title("Distribución de Precios de los inmuebles en Bogotá")
plt.xlim(50,1000)
plt.savefig("/content/drive/MyDrive/inmersion_datos/Grafico.png",format="png")
plt.show()
```
